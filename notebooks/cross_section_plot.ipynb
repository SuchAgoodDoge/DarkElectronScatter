```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
path = "../data/sigma_dist/"
files = os.listdir(path)
number_list = []
for file in files:
    #open txt file
    print(file)
    with open(path+file) as f:
        for line in f:
            if "####" in line:
                line = line.split()
                sigma = float(line[1])
                s = float(line[2])
                theta = float(line[3])
                number_list.append([sigma, s, theta])
                if len(number_list) == 100000:
                    break
    break
number_list = pd.DataFrame(number_list, columns=["sigma", "s", "theta"])
```

```python
import numpy as np
from scipy.optimize import curve_fit
#plot sigma in terms of s and theta on a 2D plot
theta_cut = number_list[number_list["theta"] > 1.45]
theta_cut = theta_cut[theta_cut["theta"] < 1.55]
plt.scatter(theta_cut["s"], theta_cut["sigma"])
plt.xlabel("s($GeV^2$)")
plt.ylabel("probability")
plt.yscale("log")
plt.savefig("../plots/report/sigma_s_bad.pdf")
```
