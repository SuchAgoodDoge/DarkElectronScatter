```python
import uproot3 as up
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

dt_ratios = ["0.33", "0.6", "2.0"]
chi_types = ["scalar", "fermion"]
```

```python
for chi_t in chi_types:
    path = f"../data/angle/{chi_t}/"
    for file in os.listdir(path):
        with up.open(path+file) as f:
            if f.keys() == []:
                continue

            print(f[b'event_tree;1'].keys())

            plt.hist(f[b'event_tree'][b'dm_px'].array())
            plt.show()
            plt.hist(f[b'event_tree'][b'dm_py'].array())
            plt.show()
            plt.hist(f[b'event_tree'][b'dm_pz'].array())
            plt.show()

            plt.hist(f[b'event_tree'][b'dm_vx'].array())
            plt.show()
            plt.hist(f[b'event_tree'][b'dm_vy'].array())
            plt.show()
            plt.hist(f[b'event_tree'][b'dm_vz'].array())
            plt.show()
```
