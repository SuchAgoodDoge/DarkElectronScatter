```python
import random as r
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#draw a sin function and a box around the 0 to pi region
x = np.linspace(0, np.pi, 100)
y = 3*np.sin(x)
plt.plot(x, y, label='f(x)')
plt.xlim(-1, np.pi+1)
plt.axvline(x=np.pi,ymin=0.05,ymax=0.95, color='r')
plt.axvline(x=0, ymin=0.05,ymax=0.95,color='r')
plt.axhline(y=0, color='black')
plt.axhline(y=3,xmin=0.195,xmax=0.805, color='r',label= '$f_{max}*$p(x)')
plt.axhline(y=1/3,xmin=0.195,xmax=0.805, color='g',label = 'p(x)')
plt.legend(loc='center right')
plt.xlabel('x')
plt.ylabel('y')
plt.savefig('../data/report/rejection_sampling.png')
plt.savefig('../data/report/rejection_sampling.pdf')
plt.show()





```
